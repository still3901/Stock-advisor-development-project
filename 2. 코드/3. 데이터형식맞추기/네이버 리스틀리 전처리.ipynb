```python
import pandas as pd


S='WOLF'
# 엑셀 파일 경로
excel_file = f'/content/drive/MyDrive/KDT/중간 프로젝트/네이버 주식토론방/{S}.xlsx'

# 엑셀 파일 불러오기
df = pd.read_excel(excel_file)

# 레이블 지우기
# drop 메서드를 사용하여 특정 열(레이블)을 제거합니다.
# axis=1은 열을 제거하겠다는 것을 의미합니다.
# 삭제할 레이블 리스트
labels_to_drop = ['LABEL-2', 'LABEL-4', 'LABEL-6', 'LABEL-7','LABEL-8']

# 레이블 삭제 (존재하지 않는 경우 무시)
df = df.drop(columns=labels_to_drop, errors='ignore')

# 열 이름 변경하기
# 딕셔너리를 사용하여 {이전 이름: 새로운 이름} 형태로 전달합니다.
df = df.rename(columns={'LABEL-1': 'head', 'LABEL-3': 'Time','LABEL-5':'Text'})

# 레이블 추가하기
# 새로운 열(레이블)을 추가하고 값을 지정합니다.
df['type'] = [f'&{S}'] * len(df)
df['src'] =['naver'] * len(df)

# 순서 변경하기
# 열(레이블)의 순서를 변경하고 싶은 경우, 열 이름의 리스트를 재배열합니다.
df = df[['Time','head','Text','type','src']]

# 날짜 형식 변환 함수 정의
def convert_date_format(date_str):
    try:
        return pd.to_datetime(date_str, format='%Y.%m.%d.').strftime('%Y-%m-%d')
    except ValueError:
        return None

# 날짜 형식 변환 적용
df['Time'] = df['Time'].apply(convert_date_format)

# 수정된 데이터프레임을 새로운 엑셀 파일로 저장
new_excel_file = f'/content/drive/MyDrive/KDT/중간 프로젝트/네이버 주식토론방/수정한거/m_{S}.xlsx'
df.to_excel(new_excel_file, index=False)  # index=False로 설정하여 인덱스를 저장하지 않습니다.
```

```python
import pandas as pd
import yfinance as yf

# 엑셀 파일 경로 및 파일 이름 설정
excel_file_path = '/content/drive/MyDrive/KDT/중간 프로젝트/네이버 주식토론방/수정한거/m_ASML.xlsx'

# 엑셀 파일 불러오기
df = pd.read_excel(excel_file_path)

# 날짜 형식을 변경하는 함수
def convert_date_format(date_str):
    return date_str.replace('.', '-')

# 'Time' 열이 있는 행에 대해만 주식 데이터를 가져와서 추가하는 함수
def get_stock_data(row):
    if 'Time' in row:
        symbol = row['type'].replace('&', '')
        date = row['Time']
        try:
            stock_data = yf.download(symbol, date, date)
            soxl_data = yf.download('SOXL', date, date)
            return pd.concat([stock_data, soxl_data.add_prefix('SOXL_')], axis=1)
        except Exception as e:
            print(f"ERROR:yfinance:[{symbol}]: {e}")
            return pd.Series()
    else:
        return pd.Series()

# 'Time' 열이 있는 행에 대해서만 주식 데이터를 가져와서 추가
additional_data = df.apply(get_stock_data, axis=1)

# 기존 DataFrame에 새로운 데이터 추가
result_df = pd.concat([df.reset_index(drop=True), additional_data.reset_index(drop=True)], axis=1)

# 결과를 새로운 엑셀 파일로 저장
result_df.to_excel('/content/drive/MyDrive/KDT/중간 프로젝트/네이버 주식토론방/수정한거/다시수정한거/31다시수정한거.xlsx', index=False)

print("새로운 엑셀 파일이 저장되었습니다.")
```
